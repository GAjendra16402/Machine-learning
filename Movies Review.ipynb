```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sentiments = SentimentIntensityAnalyzer()
```

```python
df['SentimentScore'] = df['review_content'].apply(lambda x: sentiments.polarity_scores(x)['compound'])

threshold = -0.1

df['Sentiment'] = df['SentimentScore'].apply(lambda score: 'Negative' if score < threshold else 'Not Negative')


print(df[['review_id', 'review_title', 'Sentiment']])
```

```python
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)['pos'] for i in df['Sentiment']]
df["Negative"] = [sentiments.polarity_scores(i)['neg'] for i in df['Sentiment']]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["Sentiment"]]
print(df.head())
```

```python
x = sum(df['Positive'])
y = sum(df['Negative'])
z = sum(df['Neutral'])

def sentiment_score(a,b,c):
    if (a>b) and (a>c):
        print("positive ðŸ˜˜")
    elif (b>a) and (b>c):
        print("negative ðŸ˜’")
    else:
        print('neutralðŸ‘Œ')
sentiment_score(x,y,z)
```
